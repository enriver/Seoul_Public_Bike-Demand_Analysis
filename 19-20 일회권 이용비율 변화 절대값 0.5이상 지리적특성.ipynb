```python
import pandas as pd
```

```python
daily=pd.read_csv('../dataset/ratio_daily.csv')
pop=pd.read_csv('../dataset/final_result.csv')
environ=pd.read_csv('../dataset/collected_data.csv')
```

```python
daily.head()
```

```python
daily=pd.merge(daily, pop, how='inner', on='ID')
daily=pd.merge(daily, environ, how='inner', on='ID')

daily.head()
```

```python
red_week=daily[daily['week ratio'] >= 0.5].reset_index(drop=True)
red_week.shape
```

```python
blue_week=daily[daily['week ratio'] < 0].reset_index(drop=True)
blue_week.shape
```

```python
orange_week=daily[(daily['week ratio']>=0) & (daily['week ratio'] < 0.5)].reset_index(drop=True)
orange_week.shape
```

```python
red_hol=daily[daily['hol ratio'] >= 0.5].reset_index(drop=True)
red_hol.shape
```

```python
purple_hol=daily[daily['hol ratio'] <= -0.5].reset_index(drop=True)
purple_hol.shape
```

```python
blue_hol=daily[(daily['hol ratio'] < 0) & (daily['hol ratio'] >= -0.5)].reset_index(drop=True)
blue_hol.shape
```

```python
orange_hol=daily[(daily['hol ratio']>=0) & (daily['hol ratio'] < 0.5)].reset_index(drop=True)
orange_hol.shape
```

## 월별로 구분

```python
df2019['month']=df2019['date'].dt.month
df2020_corona['month']=df2020_corona['date'].dt.month

df2019['day']=df2019['date'].dt.day
df2020_corona['day']=df2020_corona['date'].dt.day
```

```python
def two_digit(x):
    x=str(x)
    if len(x)==1:
        return '0'+x
    
    else:
        return x
```

```python
df2019['month']=df2019['month'].apply(lambda x:two_digit(x))
df2020_corona['month']=df2020_corona['month'].apply(lambda x:two_digit(x))

df2019['day']=df2019['day'].apply(lambda x:two_digit(x))
df2020_corona['day']=df2020_corona['day'].apply(lambda x:two_digit(x))
```

```python
df2019['month-day']=df2019['month']+'-'+df2019['day']
df2020_corona['month-day']=df2020_corona['month']+'-'+df2020_corona['day']
```

```python
df2020_corona.head()
```

```python
def divide_month(df):
    df_jan_mar=df[(df['month']=='01') | (df['month']=='02') | (df['month']=='03')].reset_index(drop=True)
    df_apr_jun=df[(df['month']=='04') | (df['month']=='05') | (df['month']=='06')].reset_index(drop=True)
    df_jul_sep=df[(df['month']=='07') | (df['month']=='08') | (df['month']=='09')].reset_index(drop=True)
    df_oct_dec=df[(df['month']=='10') | (df['month']=='11') | (df['month']=='12')].reset_index(drop=True)
    
    return df_jan_mar, df_apr_jun, df_jul_sep, df_oct_dec
```

```python
df2019_jan_mar, df2019_apr_jun, df2019_jul_sep, df2019_oct_dec = divide_month(df2019)
df2020_corona_jan_mar, df2020_corona_apr_jun, df2020_corona_jul_sep, df2020_corona_oct_dec = divide_month(df2020_corona)
```

```python
def show_graph(df1, df2, word):
    fig, ax1=plt.subplots()

    ax1.plot(df1['daily'], color='green', label='2019')
    ax1.plot(df2['daily'], color='blue', label='2020')
    ax1.set_ylabel('일일권 비율')
    ax1.set_ylim(0.05,0.55)

    ax2=ax1.twinx()
    ax2.bar(df2['month-day'],df2['num'], color='red', label='확진자', alpha=0.25)
    ax2.set_ylabel('확진자 수')
    ax2.set_ylim(0,600)
    ax1.set_xticklabels(labels=df2['month-day'],rotation=90)
    
    ax1.set_zorder(ax2.get_zorder()+10)
    ax1.patch.set_visible(False)

    ax1.legend(loc='upper left')
    ax2.legend()

    plt.title(word, fontsize='25', fontweight='bold', pad=20)
    plt.show()
    
```

## 거리두기 단계별 평균값 확인

20.03.22 ~ 20.05.05 (2단계)

20.08.16 ~ 20.08.30 (2단계)

20.08.31 ~ 20.09.13 (2.5단계)

20.09.14 ~ 20.10.12 (2단계)

20.10.13 ~ 20.11.23 (1단계)

20.11.24 ~ 20.12.07 (2단계)

20.12.08 ~ 20.12.31 (2.5단계)

```python
# date1~date2  구간사이의 일일권 이용비율 평균 확인 (소수점 셋째자리 반올림)

def show_mean(df,date1, date2):
    df_section=df[(df['date']>=date1) & (df['date'] <= date2)]
    
    print(round(df_section.groupby('weekday_end')['daily'].mean(),2))
```

```python
# 0322~0505

show_mean(df2020_corona, '2020-03-22', '2020-05-05')
```

```python
# 0506~0815

show_mean(df2020_corona, '2020-05-06','2020-08-15')
```

```python
# 0816~0630

show_mean(df2020_corona, '2020-08-16', '2020-08-30')
```

```python
# 0831~0913

show_mean(df2020_corona, '2020-8-31', '2020-09-13')
```

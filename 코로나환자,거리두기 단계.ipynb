# 코로나환자, 거리두기 단계에 따른 이용권 변화 추이

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open("../dataset/2019_week.pkl","rb") as fr:
    week2019 = pickle.load(fr)

with open("../dataset/2019_hol.pkl","rb") as fr:
    hol2019 = pickle.load(fr)

with open("../dataset/2020_week.pkl","rb") as fr:
    week2020 = pickle.load(fr)
    
with open("../dataset/2020_hol.pkl","rb") as fr:
    hol2020 = pickle.load(fr)
```

```python
year19=pd.concat([week2019, hol2019], ignore_index=True)
print(year19.shape)
```

```python
year20=pd.concat([week2020, hol2020], ignore_index=True)
print(year20.shape)
```

```python
year19['month']=year19['month'].apply(lambda x:str(x))
year19['day']=year19['day'].apply(lambda x:str(x))
year20['month']=year20['month'].apply(lambda x:str(x))
year20['day']=year20['day'].apply(lambda x:str(x))
```

```python
year19['month-day']=year19['month']+'-'+year19['day']
year20['month-day']=year20['month']+'-'+year20['day']
```

```python
year19.head()
```

```python
year20.head()
```

```python
data19=year19.groupby('month-day')['type'].value_counts(normalize=True)
data19
```

```python
data20=year20.groupby('month-day')['type'].value_counts(normalize=True)
data20
```

```python
set19=set(year19['month-day'])
```

```python
dict19=dict()

for i in set19:
    dict19['2019-'+i]=round(data19[i,'Daily'],2)
```

```python
df2019=pd.DataFrame.from_dict(dict19,orient='index').reset_index()
df2019.head()
```

```python
df2019.rename(columns={'index':'date',0:'daily'}, inplace=True)
df2019.head()
```

```python
df2019['date']=pd.to_datetime(df2019['date'])
df2019.head()
```

```python
df2019=df2019.sort_values(['date']).reset_index(drop=True)
df2019.head()
```

```python
df2019['dayofweek']=df2019['date'].dt.dayofweek
df2019.head()
```

## 월별로 구분

```python
df2019['month']=df2019['date'].dt.month
df2020_corona['month']=df2020_corona['date'].dt.month

df2019['day']=df2019['date'].dt.day
df2020_corona['day']=df2020_corona['date'].dt.day
```

```python
def two_digit(x):
    x=str(x)
    if len(x)==1:
        return '0'+x
    
    else:
        return x
```

```python
df2019['month']=df2019['month'].apply(lambda x:two_digit(x))
df2020_corona['month']=df2020_corona['month'].apply(lambda x:two_digit(x))

df2019['day']=df2019['day'].apply(lambda x:two_digit(x))
df2020_corona['day']=df2020_corona['day'].apply(lambda x:two_digit(x))
```

```python
df2019['month-day']=df2019['month']+'-'+df2019['day']
df2020_corona['month-day']=df2020_corona['month']+'-'+df2020_corona['day']
```

```python
df2020_corona.head()
```

```python
def divide_month(df):
    df_jan_mar=df[(df['month']=='01') | (df['month']=='02') | (df['month']=='03')].reset_index(drop=True)
    df_apr_jun=df[(df['month']=='04') | (df['month']=='05') | (df['month']=='06')].reset_index(drop=True)
    df_jul_sep=df[(df['month']=='07') | (df['month']=='08') | (df['month']=='09')].reset_index(drop=True)
    df_oct_dec=df[(df['month']=='10') | (df['month']=='11') | (df['month']=='12')].reset_index(drop=True)
    
    return df_jan_mar, df_apr_jun, df_jul_sep, df_oct_dec
```

```python
df2019_jan_mar, df2019_apr_jun, df2019_jul_sep, df2019_oct_dec = divide_month(df2019)
df2020_corona_jan_mar, df2020_corona_apr_jun, df2020_corona_jul_sep, df2020_corona_oct_dec = divide_month(df2020_corona)
```

```python
def show_graph(df1, df2, word):
    fig, ax1=plt.subplots()

    ax1.plot(df1['daily'], color='green', label='2019')
    ax1.plot(df2['daily'], color='blue', label='2020')
    ax1.set_ylabel('일일권 비율')
    ax1.set_ylim(0.05,0.55)

    ax2=ax1.twinx()
    ax2.bar(df2['month-day'],df2['num'], color='red', label='확진자', alpha=0.25)
    ax2.set_ylabel('확진자 수')
    ax2.set_ylim(0,600)
    ax1.set_xticklabels(labels=df2['month-day'],rotation=90)
    
    ax1.set_zorder(ax2.get_zorder()+10)
    ax1.patch.set_visible(False)

    ax1.legend(loc='upper left')
    ax2.legend()

    plt.title(word, fontsize='25', fontweight='bold', pad=20)
    plt.show()
    
```

## ratio 19-20   k개에 따른 군집분석 실루엣계수 확인

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# 군집 생성 후, 평균 실루엣 스코어 출력

def show_cluster(df):
  ks=range(2,6)
  inertia=list()
  #preprocessed_df=pipeline.fit_transform(df)

  for k in ks:
    model=KMeans(n_clusters=k, random_state=0).fit(df)
    inertia.append(model.inertia_)
    average_score=silhouette_score(df, model.labels_)

    print('k={} 일 때, '.format(k), end='')
    print('평균 실루엣 스코어 : {0:.3f}'.format(average_score))

  print()
  plt.plot(ks,inertia)
  plt.xlabel('numer of clusters, k')
  plt.ylabel('inertia')
  plt.xticks(ks)
  plt.show()
```

## Ratio 19 K개에 따른 군집분석 중심점 확인

```python
# k개에 따른 군집분석 중심점 확인을 위한 function 

def show_centroid(k,df):
  model=KMeans(n_clusters=k, random_state=0).fit(df)
  centers=model.cluster_centers_

  labels=model.labels_
  
  centroid_df=pd.DataFrame(centers, columns=df.columns)
  
  if k==2:
    centroid_df=centroid_df.T.reset_index().rename(columns={'index':'variables', 0:'First', 1:'Second'})
  else:
    centroid_df=centroid_df.T.reset_index().rename(columns={'index':'variables', 0:'First', 1:'Second', 2:'Third'})

  return centroid_df
```

## 연도에 따른 군집 구성 차이 확인

```python
# k개에 따른 군집 분류를 보이기 위한 function

def show_class(k, df):
  model=KMeans(n_clusters=k, random_state=0).fit(df)
  labels=model.labels_

  label_df=pd.DataFrame({'labels':labels,'station_id':df.index})

  
  first=set()
  second=set()
  third=set()
  
  for i in range(len(df)):
    if label_df['labels'].iloc[i]==0:
      first.add(label_df['station_id'].iloc[i])
    elif label_df['labels'].iloc[i]==1:
      second.add(label_df['station_id'].iloc[i])
    else:
      third.add(label_df['station_id'].iloc[i])

  print('첫번째 군집 : ',len(first),' 두번째 군집 : ',len(second),' 세번째 군집  : ',len(third))
  
  return first, second, third
```

```python
first19, second19, third19=show_class(3,ratio19_wek_reg)
third20, second20, first20=show_class(3,ratio20_wek_reg)
```

```python
print('0->0 : ',len(first19.intersection(first20)))
print('0->1 : ',len(first19.intersection(second20)))
print('0->2 : ',len(first19.intersection(third20)))
print('1->0 : ',len(second19.intersection(first20)))
print('1->1 : ',len(second19.intersection(second20)))
print('1->2 : ',len(second19.intersection(third20)))
print('2->0 : ',len(third19.intersection(first20)))
print('2->1 : ',len(third19.intersection(second20)))
print('2->2 : ',len(third19.intersection(third20)))
```

## 각 군집에 속해있는 대여소확인

```python
def make_class(k, df):
  model=KMeans(n_clusters=k, random_state=0).fit(df)
  labels=model.labels_

  label_df=pd.DataFrame({'labels':labels,'station_id':df.index})

  return label_df
```

```python
ratio19_wek_reg_cluster=make_class(3,ratio19_wek_reg)
ratio20_wek_reg_cluster=make_class(3,ratio20_wek_reg)
```

```python
ratio20_wek_reg_cluster.head()
```

```python
# 19 0->2, 1->1, 2->0 (위의 군집 중심점 그래프를 통해, 라벨 변환)
labels_transfer={0:'2', 1:'1', 2:'0'}
ratio20_wek_reg_cluster['labels']=ratio20_wek_reg_cluster['labels'].apply(lambda x:labels_transfer[x])

ratio20_wek_reg_cluster.head()
```

```python
ratio19_wek_dal_cluster=make_class(3,ratio19_wek_dal)
ratio20_wek_dal_cluster=make_class(3,ratio20_wek_dal)

# 19 0->2, 1->1, 2->0
labels_transfer={0:'2', 1:'1', 2:'0'}
ratio20_wek_dal_cluster['labels']=ratio20_wek_dal_cluster['labels'].apply(lambda x:labels_transfer[x])
```

```python
ratio19_hol_dal_cluster=make_class(2,ratio19_hol_dal)
ratio20_hol_dal_cluster=make_class(2,ratio20_hol_dal)
```

```python
ratio19_hol_reg_cluster=make_class(2,ratio19_hol_reg)
ratio20_hol_reg_cluster=make_class(2,ratio20_hol_reg)

# 19 0->1, 1->0
labels_transfer={0:'1', 1:'0'}
ratio20_hol_reg_cluster['labels']=ratio20_hol_reg_cluster['labels'].apply(lambda x:labels_transfer[x])
```

#### 연/월/일/요일 분류

```python
def year_month_day(df):
    df['year']=df['rent_time'].dt.year
    df['month']=df['rent_time'].dt.month
    df['day']=df['rent_time'].dt.day
    df['dayofweek']=df['rent_time'].dt.dayofweek

    weeks={0:'Mon',
          1:'Tue',
          2:'Wed',
          3:'Thu',
          4:'Fri',
          5:'Sat',
          6:'Sun'}

    df['dayofweek']=df['dayofweek'].map(weeks)
    
    return df
```

```python
df2019=year_month_day(df2019)
df2020=year_month_day(df2020)
```

```python
from scipy.stats import mode

# 데이터 크기, 최빈값, 중앙값, 평균 확인
def show_describe(data_name, data):
    print("*****"+data_name+"["+str(data.name)+"]*****")
    print("데이터 크기 : {0:,.2f}".format(len(data)))
    print("최빈값 : {0:,.2f}".format(mode(data)[0][0]))
    print("중앙값 : {0:,.2f}".format(data.quantile(.5)))
    print("평균 : {0:,.2f}".format(data.mean()))
    print('최대값 : {0:,.2f}'.format(data.max()))
```

```python
show_describe('regular19',regular19['travel_time'])
```

```python
show_describe('daily19',daily19['travel_time'])
```

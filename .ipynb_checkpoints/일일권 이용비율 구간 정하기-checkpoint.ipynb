## 일회권 분포 확인

```python
plt.rcParams['font.family']='Hancom Gothic'

plt.figure(figsize=(8,6))
sns.distplot(day['week ratio'], label='주중')
sns.distplot(day['hol ratio'], label='주말')
plt.title('일회권 분포 확인', fontsize=14)
plt.legend()
plt.show()
```

```python
def show_criteria(df):
    mean_value=round(df.mean(),2)
    median_value=round(df.median(),2)
    print('평균 : ',mean_value)
    print('중앙값 : ',median_value)
    
    return mean_value, median_value
```

```python
week_mean, week_median=show_criteria(day['week ratio'])
```

```python
hol_mean, hol_median=show_criteria(day['hol ratio'])
```

### 평균을 기준으로 나눔

```python
week_under_zero=day[day['week ratio']<0].reset_index(drop=True)
week_under_mean=day[(day['week ratio']>=0) & (day['week ratio'] < week_mean)].reset_index(drop=True)
week_over_mean=day[day['week ratio']>=week_mean].reset_index(drop=True)
```

```python
print(week_under_zero.shape)
print(week_under_mean.shape)
print(week_over_mean.shape)
```

```python
hol_under_zero=day[day['hol ratio']<0].reset_index(drop=True)
hol_under_mean=day[(day['hol ratio']>=0) & (day['hol ratio'] < hol_mean)].reset_index(drop=True)
hol_over_mean=day[day['hol ratio']>=hol_mean].reset_index(drop=True)
```

```python
print(hol_under_zero.shape)
print(hol_under_mean.shape)
print(hol_over_mean.shape)
```

### 중앙값을 기준으로 나눔

```python
week_under_med=day[(day['week ratio']>=0) & (day['week ratio'] < week_median)].reset_index(drop=True)
week_over_med=day[day['week ratio']>=week_median].reset_index(drop=True)
```

```python
print(week_under_med.shape)
print(week_over_med.shape)
```

```python
hol_under_med=day[(day['hol ratio']>=0) & (day['hol ratio'] < hol_median)].reset_index(drop=True)
hol_over_med=day[day['hol ratio']>=hol_median].reset_index(drop=True)
```

```python
print(hol_under_med.shape)
print(hol_over_med.shape)
```

### 나눈 결과를 기준으로 주변환경 데이터 & 인구 총계 데이터 확인

```python
def make_describe(df1,df2,df3,word):
    dataframes=[df1,df2,df3]
    
    for df_order in range(len(dataframes)):
        df=dataframes[df_order]

        mean_dict=dict()
        columns=df.columns[3:]

        for i in range(len(columns)):
            mean_dict[columns[i]+word]=df[columns[i]].mean()

        if df_order==0:
            first_df=pd.DataFrame.from_dict(mean_dict, orient='index').T
        else:
            new_df=pd.DataFrame.from_dict(mean_dict, orient='index').T
            first_df=pd.concat([first_df,new_df], ignore_index=True)
            
    return first_df
```

```python
week_mean_desc=make_describe(week_under_zero, week_under_mean, week_over_mean, 'mean')
week_mean_desc=week_mean_desc.rename(index={0:'week_under_zero', 1:'week_under_mean', 2:'week_over_mean'})
week_mean_desc.head()
```

```python
hol_mean_desc=make_describe(hol_under_zero, hol_under_mean, hol_over_mean, 'mean')
hol_mean_desc=hol_mean_desc.rename(index={0:'hol_under_zero', 1:'hol_under_mean', 2:'hol_over_mean'})
hol_mean_desc.head()
```

```python
week_50_desc=make_describe(week_under_zero, week_under_med, week_over_med, 'median')
week_50_desc=week_50_desc.rename(index={0:'week_under_zero', 1:'week_under_50', 2:'week_over_50'})
week_50_desc.head()
```

```python
hol_50_desc=make_describe(hol_under_zero, hol_under_med, hol_over_med, 'median')
hol_50_desc=hol_50_desc.rename(index={0:'hol_under_zero', 1:'hol_under_50', 2:'hol_over_50'})
hol_50_desc.head()
```

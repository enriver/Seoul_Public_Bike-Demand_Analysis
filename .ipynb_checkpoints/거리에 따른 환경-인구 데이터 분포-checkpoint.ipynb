### 평균, 중앙값 확인

```python
def show_criteria(df):
    mean_value=round(df.mean(),2)
    median_value=round(df.median(),2)
    print('평균 : ',mean_value)
    print('중앙값 : ',median_value)
    
    return mean_value, median_value
```

## 나눈 결과를 기준으로 주변환경 데이터 & 인구통계 데이터 분포 확인

```python
def make_describe(df1,df2,df3,word):
    dataframes=[df1,df2,df3]
    
    for df_order in range(len(dataframes)):
        df=dataframes[df_order]

        mean_dict=dict()
        columns=df.columns[3:]

        for i in range(len(columns)):
            mean_dict[columns[i]+'_'+word]=round(df[columns[i]].mean(),1)

        if df_order==0:
            first_df=pd.DataFrame.from_dict(mean_dict, orient='index').T
        else:
            new_df=pd.DataFrame.from_dict(mean_dict, orient='index').T
            first_df=pd.concat([first_df,new_df], ignore_index=True)
            
    return first_df
```

### 평균

```python
week_mean_desc250=make_describe(week_under_zero250, week_under_mean250, week_over_mean250, 'mean')
week_mean_desc250=week_mean_desc250.rename(index={0:'week_under_zero', 1:'week_under_mean', 2:'week_over_mean'})
print('250m 주중 평균')
week_mean_desc250.head()
```

```python
hol_mean_desc250=make_describe(hol_under_zero250, hol_under_mean250, hol_over_mean250, 'mean')
hol_mean_desc250=hol_mean_desc250.rename(index={0:'hol_under_zero', 1:'hol_under_mean', 2:'hol_over_mean'})
print('250m 주말 평균')
hol_mean_desc250.head()
```
